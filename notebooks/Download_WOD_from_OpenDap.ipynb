```python
import xarray
import numpy as np
import pandas 
from matplotlib import pyplot as plt
import os
```

```python
start_year = 2001
end_year   = 2023

platforms_to_get = ['ctd','xbt','pfl','mrb','gld','osd','uor','apb']


base_url    = 'https://www.ncei.noaa.gov/thredds-ocean/dodsC/ncei/wod/'
output_path = '/oa-decadal-climate/work/observations/WOD2018'



for i_year in range(start_year, end_year):
    for i_platform in platforms_to_get:
        print('Getting file: ', 'wod_' + i_platform + '_' +  str(i_year) + '.nc')
        current_url = os.path.join(base_url,str(i_year),'wod_' + i_platform + '_' +  str(i_year) + '.nc')
        
        try: 
            WOD_dataset = xarray.open_dataset(current_url + '#fillmismatch',decode_times=True)
        
            output_file_name =  'wod_' + i_platform + '_' + str(i_year) + '.nc'
        
            print('Writing file:', output_file_name)
            if not os.path.isdir(os.path.join(output_path,str(i_year))):
                os.mkdir(os.path.join(output_path,str(i_year)))
    
            WOD_dataset.to_netcdf(os.path.join(output_path,str(i_year),output_file_name)) #,encoding=encoding) #,engine='netcdf4')
        except: 
            print('Cannot file file: ', current_url)
```
